Importing required library and dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df
```

```python
ax = sns.scatterplot(data = df, 
                x = 'workout_type', 
                y = 'time',
                size = 'calories',
                hue = 'workout_type',
                legend = 'brief')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1.25,1))
```

```python
sns.set(rc={"figure.figsize":(10, 10)})
ax = sns.scatterplot(data = df, 
                x = 'distance', 
                y = 'time',             
                size = 'distance',
                hue = 'workout_type',
                x_bins = "avg_speed")
sns.move_legend(ax, "upper right", bbox_to_anchor=(1,1))
```
