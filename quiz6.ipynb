Pulling out the dataset from googleclode using bigquery than joining the bikeshare_trips and bikeshare_stations on the start_station_name. than  computing the average time on trips starting from 'parkland' and from 'sidewalks.


```python
%%bigquery --project data-visualization-338800
WITH t as(
SELECT name, property_type
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
)
SELECT t.property_type as stating_startion_type, 
    AVG (trips.duration_minutes) as average_min,
    COUNT(*) as number_of_trips,
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trips
  JOIN t
  ON trips.start_station_name = t.name

GROUP BY t.property_type
ORDER by average_min
```

Output shows the starting_startion_type and average time on trips starting from 'parkland' and from 'sidewalks.

According to the result from the table we can say that the  average time on trips starting from 'parkland' is longer than from 'sidewalks.
