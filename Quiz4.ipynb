Input2 = Pulling out the dataset from googleclode using bigquery and also setting the range of data to 5.

```python
%%bigquery --project data-visualization-338800
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

Output2 = Displaying the dataset with only 5 sets of data.

Input3 = Computing the average time for a trip based on starting point from the dataset.

```python
%%bigquery --project data-visualization-338800
SELECT 
  start_station_name, 
  AVG(duration_minutes) as Average_time_for_trip
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name is not null 
GROUP BY start_station_name
```

Output3 = Displaying the list of the start station name with its average time for trip.

Input4 = Computing the number of trips starting at each of starting point.

```python
%%bigquery --project data-visualization-338800
SELECT 
  start_station_name,
  COUNT(*) as total_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY total_trips DESC
```

Output4 = Displaying the list of the start station name  with their total number of trips.
