Importig the necessary libraries for the project.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

Requesting data from the wikipedia page.

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_parse = r.text
html_soup = BeautifulSoup(html_parse,'lxml')
```

Calculating the number of the table in the list.

```python
len(html_soup.find_all('table'))
```

Output shows there are 7 tables in the list.

Computing the table from our wikipedia page.

```python
marvelTable = html_soup.find_all('table', class_ = 'wikitable plainrowheaders')
dataFrame = pd.read_html(str(marvelTable))[0]
dataFrame
```

Here,

 i will be removing the unnecessary rows over the heading using the droplevel method in pandas.

```python
dataFrame.columns = dataFrame.columns.droplevel(1)
dataFrame
```

According to the table the row 6 & 13 are not movies. so i will be removing these two rows and rearrange the order of rows.

```python
dataFrame = dataFrame.drop(dataFrame.index[[6,13]])
dataFrame
```

Output shows the table after the movie is removed, however the row number is not correct.

computing a new table by Rearranging the row number.

```python
dataFrame = dataFrame.reset_index(drop=True, inplace=False)
dataFrame
```

Now removing the Unnamed: 5_level_0 from the index as it doesnot hold any informative value.

```python
del dataFrame['Unnamed: 5_level_0']
dataFrame
```

Output shows the table without the unnecessary Unnamed: 5_level_0 column.
