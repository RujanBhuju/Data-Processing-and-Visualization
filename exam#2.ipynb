(A) Here, i will be Storing the data as a pandas dataframe. I will also Examine each datatype and comment on the appropriateness of each.

Importing the required library.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

examining the datatypes.

```python
df.dtypes
```

(B) Removing the column that repeats the indexes and is ‘unnamed’ as a column.


```python
df=df.drop(columns=["Unnamed: 0"])
df.columns
```

(C) Cleaning the column names to remove the unit declaration, (%), using regular expressions. and also The column name should not have any trailing spaces after cleaning.

```python
df.columns=df.columns.str.replace('[%,(,)]', '')
df.columns=df.columns.str.strip()
df.columns
```

(D) Converting ‘activity day’ column into a datetime format.

```python
df.activity_day=pd.to_datetime(df.activity_day)
df.head()
```

(E) Imputing ‘total steps’ by replacing the ‘NaN’s with an appropriate number of steps. and also Converting to appropriate datatype.

Using fillna to replace the 'NaN's with appropriate number of steps.

```python
df.total_steps=df.total_steps.fillna(0)
df.total_steps=df.total_steps.astype(int)
df.head(10)
```

The output shows that the N/A is replaced by 0 where there are no total steps.

#(F) Converting non-empty ‘avg pace’ into a float that is still representative of the information contained in the column.and Recalling that there are 60 seconds in one minute so 3:30 is equivalent to 3.5 minutes.

splitting the string from ":" to seperate hour and minute

```python
date_format= df[df.avg_pace.notnull()]["avg_pace"].str.split(':')
```

```python
hour,min = date_format.str[0], date_format.str[1]  
df.loc[df.avg_pace.notnull(),"avg_pace"]=hour.astype(float) + (min.astype(float) / 60)
```

Checking the final data type.

```python
type(df.avg_pace[3])
```

(G) Grouping data by ‘workout type’ and find the mean, median, count and standard deviation of ‘calories’.

```python
df.groupby("workout_type")["calories"].agg({"mean","median","count","std"})
```

(H) Creating an indicator column that identifies if the activity achieved 30% or more ‘aerobic’ activity.

first assuming that all the values are false and than setting up the required condition for true values.
