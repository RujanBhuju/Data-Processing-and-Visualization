Importing the required library.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

Importing data from  Air B&B NYC data.

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
df.head()
```

i will be creating bar graph of the data using  plot function. for the graph i will be using xlabel for x-axis to represent neighbourhood and ylabel for yaxis to represent price.

```python
df.groupby('neighbourhood_group').price.agg('max').plot(kind = 'bar', title = 'Max Price', xlabel = 'neighbourhood', ylabel = 'price')
```

showing the bar horizontally for better visuability.

```python
df.groupby('neighbourhood_group').price.agg('mean').plot(kind='barh', title = 'Neighbour by Price', xlabel= "Neighbourhood", ylabel = 'Price')
```

```python
df.groupby(['neighbourhood_group','room_type']).price.agg('max').plot(kind = 'barh',title = 'Max price', ylabel= 'Price in dollar')
```

```python
df.groupby(['room_type', 'neighbourhood_group'])['price'].mean().plot(kind = 'bar')
```

```python
ax = df.groupby('room_type').agg('mean').plot(kind = 'bar', ylim =[0,1000])

for container in ax.containers:
    ax.bar_label(container)
```

```python
plt.figure(figsize=(10,10))
df1 = sns.barplot(data=df, x='room_type',y='price') 
```

```python
plt.figure(figsize=(10,10))
df2 = sns.barplot(data=df1, x='price',y='room_type')
```
