7) Creating ‘Goals’ and ‘Assists’ as scatter. Signifying my favorite club in some way on the scatterplot.


```python
sns.relplot(data=df, x = 'Goals', y= 'Assists' )
plt.title('A scatterplot of goals and assists')
```

The output shows scatterplot of all data for the goals and assist in the dataset.
first of all i will be creating a new column in the dataset where I can store the sum of goals and assist of all the Real Madrid players from each row.
second, i will be creating a function that will take in a row of data and return the sum of goals and assist from that row. Then i will use str.contains function to get only rows that contains Real Madrid from Club Column. Finally, I will use Lambda to pass those rows into my function which return the sum and put it into a new column Real_Madrid.

```python
def RealMadridGoals(row):
  return row.Goals

def RealMadridAssists(row):
  return row.Assists
```

```python
df2 = df[df['club'].str.contains('Real Madrid')]

df['RMFC_goals'] = df2.apply(lambda row: RealMadridGoals(row), axis=1)

df['RMFC_assists'] = df2.apply(lambda row: RealMadridAssists(row), axis=1)
```

Now Real Madrid players have their tally but other players who does not play for Real Madrid now will have NULL values(0). I will be passing NULL value (0) using fillna function from pandas library. Also, to remove decimal i will be converting the column into integer.

```python
df['RMFC_goals'] = df['RMFC_goals'].fillna(0).astype(int)
df['RMFC_goals'] = df['RMFC_goals'].astype('int')
df['RMFC_assists'] = df['RMFC_assists'].fillna(0).astype(int)
df['RMFC_assists'] = df['RMFC_assists'].astype('int')
```
