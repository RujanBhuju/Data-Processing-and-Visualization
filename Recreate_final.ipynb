Importing the required library.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

Importing the required data which i have uploaded on my github. 

```python
df = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')
df.head()
```

```python
x_values = df.index
```

```python
df['date'] = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')['DATE']
df['unrate'] = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')['UNRATE']
df['CPIAUCSL_PC1'] = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')['CPIAUCSL_PC1']
```

```python
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['unrate'],marker='',color='#457DEE',label ='Unemployment Rate')
plt.plot(df['date'],df['CPIAUCSL_PC1'],marker='',color='#840B0B',label='Consumer Price Index for All Urban Consumers: All items in U.S City Average')
plt.legend(loc="upper center")
```
