Importing the required library.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox
```

Importing the required data which i have uploaded on my github. 

The Source of my dataset is U.S. Bureau of Labor Statistics.

```python
df = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')
df.head(10)
```

```python
x_values = df.index
```

Plotting the chart.

```python
df['date'] = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')['DATE']
df['unrate'] = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')['UNRATE']
df['CPIAUCSL_PC1'] = pd.read_csv('https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/fredgraph.csv')['CPIAUCSL_PC1']
```

```python

plt.figure(figsize=(15,10))
plt.plot(df['date'],df['unrate'],marker='',color='#457DEE',label ='Unemployment Rate')
plt.plot(df['date'],df['CPIAUCSL_PC1'],marker='',color='#840B0B',label='Consumer Price Index for All Urban Consumers: All items in U.S City Average')
plt.legend(loc="upper center")
```

Adding the logo on the left top and the black line at the zero. i will also be adjusting the dates on the x-axis.

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)     
ax.set_xticks(np.arange(1950,2020,10))
ax.plot(df['date'],df['unrate'],marker='',color='#457DEE',label ='Unemployment Rate')
ax.plot(df['date'],df['CPIAUCSL_PC1'],marker='',color='#840B0B',label='Consumer Price Index for All Urban Consumers: All items in U.S City Average')
data_str = '1950                       1960                      1970                       1980                       1990                         2000                                2010                                     2020'
data_str2 = 'Percent, Percent Change from Year Ago'
ax.set_xlabel(data_str,labelpad=6.0)
ax.set_ylabel(data_str2,labelpad=6.0)
plt.legend(loc="upper center")


plt.axhline(y = 0.0, color = '#000000', linestyle = '-')


link = 'https://raw.githubusercontent.com/RujanBhuju/Data-Processing-and-Visualization/main/flogo.png'
arr_lena = plt.imread(link)
imagebox = OffsetImage(arr_lena, zoom=0.3)
ab = AnnotationBbox(imagebox, (135, 14.7), frameon = False)
ax.add_artist(ab)
plt.show()

plt.show()
plt.show()
```
