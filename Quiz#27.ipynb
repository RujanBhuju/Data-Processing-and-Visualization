I will also Annotate the graph in some way and Explain what a Mosaic Plot might be able to show us that a pie and bar chart cannot.

Importing the required library and dataset.

```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

```python
df.groupby('workout_type').workout_type.agg('count')
```

Changing the colors

```python
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, autopct='%.0f%%')
plt.show()
```

```python
colors = sns.color_palette('bright')[0:12]
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, autopct='%.0f%%', colors=colors)
plt.show()
```

```python
colors = sns.color_palette('bright')
plt.pie(x=df.groupby('workout_type').workout_type.agg('count'),labels = df.groupby('workout_type').workout_type.agg('count').index, autopct='%.0f%%', colors=colors[1:11]) #this is changing the color using the 1:10 ratio. It is changing the color but not assigning the unique colr. But it still helped to differentiate those two together.
plt.show()
```

Mosaic plots helps us show relationships and gives a visual way to compare groups. Mosiac plots are able to show relation between two categorial variables.It allows us to group sizes by scaling in x-axis which is easier to view and compare. thus Mosaic Plot might be able to show us that a pie and bar chart cannot.
